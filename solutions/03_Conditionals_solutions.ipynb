### Comparison Operators:

More often than not, your boolean variables will be a result of statements containing comparison operators.

| Symbol |    Operation   | Usage | Boolean Outcome |
|:------:|:--------------:|:-----:|:-------:|
|    ==   |  is equal to  |`10==5*2`| True |
|    !=   | is not equal to | `10!=5*2` | False |
|    >  | Greater than |  `10 > 2` | True |
|    <   |    Less than    |  `10 < 2` | False |
| >= | Greater than _or_ equal to | `10 >= 10` | True |
| <= | Less than _or_ equal to | `10 <= 10` | True |

#### Other operators:

| "Symbol" |    Operation   | Usage | Boolean Outcome |
|:------:|:--------------:|:-----:|:-------:|
| `is` | identity | 10 is 10 | True |
| `in` | membership | 10 in [1, 2, 3] | False |

But you'll see `is` and `in` described in later notebooks.

## Conditional Statements
**Conditionals** are statements that check if a condition is met, using the `if` statement, and then only execute a set of code if the condition evaluates as `True`.

- `if`
- `elif` (else if): After an `if`, you can use `elif` statements to check additional conditions.
- `else`: After an `if`, you can use an `else` that will run if the conditional(s) above have not run.

### If/elif/else syntax
- Indentation matters! Your statements in the `if` block need to be indented by a tab or four spaces.
- You need a colon after `if`, `elif`, and `else`

```python
condition = False

if condition:
    print('This code executes if the condition evaluates as True.')

else:
    print('This code executes if the condition evaluates as False')
```

#### Imagine we're recording neural activity from a motion-selective area in the brain:

#### The neuron we are recording from responds strongly to upward motion: 90Â° in a circular space. This neuron also responds to directions close to 90Â°, but weaker when compared to a response to 90Â° motion.


```python
neuron_spike_count = 0

# We know its preferred motion direction
strongly_preferred_direction = 90
preferred_direction_minimum = strongly_preferred_direction - 10
preferred_direction_maximum = strongly_preferred_direction + 10

# And the current stimulus motion direction -- you can change this!
current_direction = 75

# Simulate neuron spike count
if current_direction == strongly_preferred_direction: # strongly preferred motion direction
    neuron_spike_count = neuron_spike_count + 8

elif (current_direction >= preferred_direction_minimum) and (current_direction <= preferred_direction_maximum): # weakly preferred motion direction
    neuron_spike_count += 3

else:
    print('The neuron is not sensitive to this motion direction!')


print('Neuron spike count: ' + str(neuron_spike_count))
```

### Properties of Conditionals
- Conditionals can take any expression that can be evaluated as `True` or `False`.
- The order of execution for conditional blocks is *always* `if` then `elif`(s) then `else`.
- If the `elif` is at the end, it will never be tested, as the `else` will have already returned a value once reached (and Python will throw an error).
- An `else` statement is not required, but if both the `if` and the `elif` condtions are not met (both evaluate as `False`), then nothing is returned.
- **At most one component (`if` / `elif` / `else`) of a conditional will run**, which is why the order of your conditional statements is *crucial*.
